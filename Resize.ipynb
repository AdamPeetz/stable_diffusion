```python
# import libraries
import os
import cv2
from tqdm import tqdm

# create count variable
i = 0
dims = 128

# define file path for processing
image_path = "C://Users//Adamg//Desktop//generators//crystal_nova_voyager//"
save_path = "C://Users//Adamg//Desktop//generators//crystal_nova_voyager_resized//"

# iterate through files in directory and rename to sequential numbers
for file in tqdm(os.listdir(image_path)):
    
    # read image to process
    in_process_image = cv2.imread(os.path.join(image_path,file))  
    width_defined_image = cv2.resize(in_process_image, (dims,dims))
    name = "%05d.jpg" % (i)
    target = save_path + name
    cv2.imwrite(target,width_defined_image)
    i+=1
        
```
